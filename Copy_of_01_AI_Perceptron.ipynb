```python
import numpy as np

class Perceptron:
    def __init__(self, input_size, learning_rate=0.1, epochs=100):
        self.weights = np.zeros(input_size + 1)  # Includes one bias weight
        self.learning_rate = learning_rate
        self.epochs = epochs

    def activation_function(self, x):
        return 1 if x >= 0 else 0

    def predict(self, x):
        z = np.dot(x, self.weights[1:]) + self.weights[0]
        return self.activation_function(z)

    def train(self, X, y):
        for epoch in range(self.epochs):
            for i in range(len(X)):
                x_i = X[i]
                y_pred = self.predict(x_i)
                error = y[i] - y_pred
                # Update weights and bias
                self.weights[1:] += self.learning_rate * error * x_i
                self.weights[0] += self.learning_rate * error

# Sample Dataset: OR Gate
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])
y = np.array([0, 1, 1, 1])  # OR Gate outputs

# Initialize and train perceptron
perceptron = Perceptron(input_size=2, learning_rate=0.1, epochs=10)
perceptron.train(X, y)

# Test the perceptron
print("Testing the Perceptron:")
for x in X:
    print(f"Input: {x}, Predicted Output: {perceptron.predict(x)}")
```

Frank Rosenblatt developed the Perceptron in 1958, marking a foundational moment in the history of artificial intelligence and neural networks. The Perceptron was designed as an early artificial neural network capable of pattern recognition through supervised learning. It functioned based on a two-layer architecture:
	1.	Input Layer: Accepted input signals (data points) and processed them.
	2.	Output Layer: Made decisions or classifications by applying weights and thresholds to the inputs.

Key Features of the Perceptron

	•	Learning Rule: Rosenblatt introduced a learning rule to adjust weights iteratively, ensuring the network could correctly classify input data after multiple training cycles.
	•	Linear Separability: The Perceptron could classify data into two categories if the data was linearly separable (separable by a straight line in two dimensions or a hyperplane in higher dimensions).

Limitations

The Perceptron faced criticism in 1969, especially through Marvin Minsky and Seymour Papert’s book Perceptrons, which highlighted its inability to solve problems requiring non-linear separation (e.g., the XOR problem). This critique caused a slowdown in neural network research, known as the AI Winter.

Legacy

Despite its limitations, the Perceptron inspired further research into multi-layer neural networks (e.g., backpropagation algorithms) and paved the way for modern deep learning. Today, its principles are reflected in advanced AI systems, demonstrating Rosenblatt’s profound impact on the field.

```python
import numpy as np
import matplotlib.pyplot as plt

class Perceptron:
    def __init__(self, input_size, learning_rate=0.1, epochs=100):
        self.weights = np.zeros(input_size + 1)  # Includes one bias weight
        self.learning_rate = learning_rate
        self.epochs = epochs

    def activation_function(self, x):
        return 1 if x >= 0 else 0

    def predict(self, x):
        z = np.dot(x, self.weights[1:]) + self.weights[0]
        return self.activation_function(z)

    def train(self, X, y):
        for epoch in range(self.epochs):
            for i in range(len(X)):
                x_i = X[i]
                y_pred = self.predict(x_i)
                error = y[i] - y_pred
                # Update weights and bias
                self.weights[1:] += self.learning_rate * error * x_i
                self.weights[0] += self.learning_rate * error

    def plot_decision_boundary(self, X, y):
        # Plot data points
        for i in range(len(X)):
            if y[i] == 0:
                plt.scatter(X[i][0], X[i][1], color='green', marker='o', label='Not Spam' if i == 0 else "")
            else:
                plt.scatter(X[i][0], X[i][1], color='red', marker='x', label='Spam' if i == 0 else "")

        # Plot decision boundary
        x_min, x_max = -0.5, 1.5
        y_min, y_max = -0.5, 1.5
        xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
        Z = np.array([self.predict([x, y]) for x, y in zip(xx.ravel(), yy.ravel())])
        Z = Z.reshape(xx.shape)

        plt.contourf(xx, yy, Z, alpha=0.2, cmap='coolwarm')
        plt.title("Spam Email Classification Decision Boundary")
        plt.xlabel("Contains 'offer'")
        plt.ylabel("Email Length > 100")
        plt.legend()
        plt.show()


# Simplified dataset
# Feature 1: Contains "offer" (0 or 1)
# Feature 2: Email Length > 100 (0 or 1)
X = np.array([
    [0, 0],  # Not Spam
    [0, 1],  # Not Spam
    [1, 0],  # Spam
    [1, 1]   # Spam
])

# Labels (0 = Not Spam, 1 = Spam)
y = np.array([0, 0, 1, 1])

# Initialize and train perceptron
perceptron = Perceptron(input_size=2, learning_rate=0.1, epochs=10)
perceptron.train(X, y)

# Plot decision boundary
perceptron.plot_decision_boundary(X, y)

# Test with a new email
new_email = np.array([1, 0])  # Contains "offer", short email
print(f"New email classified as: {'Spam' if perceptron.predict(new_email) == 1 else 'Not Spam'}")
```
